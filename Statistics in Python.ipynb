# OUTLIERS


```python
 data = [11, 12, 14, 12, 13, 15, 11, 30, 10, 100, 12, 14, 11, 9, 200]
plt.hist(data)
```

```python
outlier=[]
def detect_outliers(data):
    mean=np.mean(data)
    std=np.std(data)
    for i in data:
        Z=i-mean/std
        if(np.abs(Z)>3):
            outlier.append(i)
    return outlier
```

```python
detect_outliers(data)
```
